```python
import pandas as pd, numpy as np
import os

from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
df = pd.read_csv('/Users/leeminho/Downloads/TT_comment.csv')
df
```

```python
names = np.unique(df['name'].values)
names
```

```python
del df['Unnamed: 0']
del df['Unnamed: 0.1']
```

```python
total_df = pd.DataFrame()
for i in range(len(names)):
    one_df = df.loc[df['name'] == names[i]]
    dict = {}
    for idx in range(len(df)):
        ex = okt.nouns(df.loc[idx, 'comment'])
    
        for i in range(len(ex)):
            if ex[i] in dict:
                dict[ex[i]] += 1
            else:
                dict[ex[i]] = 1
                
    player = pd.DataFrame(dict.items())
    player = player.sort_values(by = [1], axis = 0, ascending = False)
    
    player_list = []
    for j in range(len(player[0].values)):
        for k in range(len(names)):
            if player[0].values[j] in names[k]:
                if len(player[0].values[j]) > 1:
                    player_list.append(names[k])
    if len(player_list) > 5:
        player_list = player_list[:5]
    player_list = [player_list]
    one_df = pd.DataFrame({'name': names[i], 'player': player_list})
    total_df = pd.concat([total_df, one_df])

total_df
        
    
            
```
